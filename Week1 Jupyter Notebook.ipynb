```python
dog_breeds = ['American Pit Bull Terrier', 'German Shepherd Dog', 'Chow Chow', 'Cocker Spaniel', 'Australian Cattle Dog', 
        'Labrador Retriever','American Staffordshire Terrier', 'Supermutt']
percentage = [20.1, 16.9, 16.3, 12.6, 9.6, 8.4, 4.2, 11.9]

skylar_breeds_df = pd.DataFrame(list(zip(dog_breeds,percentage)), columns = ['Breed','Percentage'])

skylar_breeds_df.plot(kind = 'bar', x = 'Breed', y ='Percentage', title = 'My Dog Skylar')
plt.show()
```
